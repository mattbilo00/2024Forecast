```python
import requests
import pandas as pd

## imports my api key
from my_info import *
var_split = "!!"
```

```python
## Find a list of all state FIPS codes and format
path = '/Users/mattg/Personal Projects/2024_forecast'
fips_df = pd.read_csv(path + "/fips_info.csv")
fips_df['StateFIPS'] = fips_df['StateFIPS'].astype(int).apply(lambda x: f"{x:02d}")
state_fips = fips_df['StateFIPS'].unique().tolist()
print(state_fips)
```

```python
# load the race variables
race_variables = ''
var_int = 78

for i in range(6):
    race_variables += 'DP1_00' + str(var_int) + 'P,'
    if i == 5:
        race_variables += 'DP1_00' + str(var_int) + 'C'
    else:
        race_variables += 'DP1_00' + str(var_int) + 'C,'
    var_int += 1
    
print(race_variables)
```

```python
# Specify the dataset code for the Demographic and Housing Characteristics dataset
dataset_code = 'dp'

# Specify the API endpoint to get variable information
base_url = f'https://api.census.gov/data/2020/dec/{dataset_code}/variables.json?key={api_key}'

# Make an API request to get variable information
response = requests.get(base_url)

# Check for errors
race_var_dict = {}
if response.status_code != 200:
    print(f"Error: {response.status_code}")
else:
    variable_data = response.json()
    # Print variable names and labels
    for variable, info in variable_data['variables'].items():
        if variable in race_variables:
            label_parts = info['label'].split(var_split)
            race_var_dict[variable] = label_parts[4]
        
print(race_var_dict)
```

```python
"""
Print race population numbers for all states
Go to following link for race variable info:
https://api.census.gov/data/2020/dec/dp/variables.html
"""

base_url = 'https://api.census.gov/data/2020/dec/dp'
f = open(path + "/race_by_state.txt","w+")

for fips_cd in state_fips:
    url = f'{base_url}?get={race_variables}&for=state:{fips_cd}&key={api_key}'
    response = requests.get(url)
    
    # Check for errors
    if response.status_code != 200:
        print(f"Error: {response.status_code}")
    else:
        data = response.json()
        # Extract headers and values
        headers = data[0]
        headers.pop()
        values = data[1]

        for i in range(0, len(headers), 2):
            print(f'"{fips_cd}_{headers[i]}": [{values[i]}, {values[i+1]}, "{race_var_dict[headers[i]]}"],')
            f.write(f'"{fips_cd}_{headers[i]}": [{values[i]}, {values[i+1]}, "{race_var_dict[headers[i]]}"],\n')
            
f.close()
```
